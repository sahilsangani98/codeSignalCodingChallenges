```
def shapeArea(n):
    ans = ((n-1)*(n-1)) + ((n)*(n))
    #  (n * n-1) -1 
    # n* n - 
    # 2*1 - 1 -> 1
    # 3*2 - 1 --> 5
    # 4*3 - 1 --> 11. -----13
    return ans
```
